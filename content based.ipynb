###  Content Based Model (Movie ratings Prediction)

```python
import pandas as pd
import numpy as np
```

```python
column_names=['user_id','item_id','rating','timestamp']
```

```python
df=pd.read_csv("D://Files//file.tsv",sep='\t',names=column_names)
```

```python
df.head()
```

```python
movie_titles=pd.read_csv("D://Files//Movie_Id_Titles.csv")
```

```python
movie_titles.head(2)
```

```python
data=pd.merge(movie_titles,df,on='item_id')
```

```python
data.head()
```

```python
data.groupby('title')['rating'].mean().sort_values(ascending=False)
```

```python
# count rating of all movies
data.groupby('title')['rating'].count().sort_values(ascending=False)
```

```python
# creating data frame with count rating values
ratings=pd.DataFrame(data.groupby('title')['rating'].mean())
ratings['num_of_ratings']=pd.DataFrame(data.groupby('title')['rating'].count())
                     
```

```python
ratings.head()
```

```python
# create user-item matrix ( moviematrix)
moviemat=data.pivot_table(index='user_id', columns='title',values='rating')
```

```python
moviemat.head(2)
```

```python
# which are users watched the movie star wars and votes (ratings)
starwar=moviemat['Star Wars (1977)']
#starwar.dropna(inplace=True)
```

```python
starwar
```

```python
# Find similarity
similar=moviemat.corrwith(starwar)
correl=pd.DataFrame(similar,columns=['correlation'])
correl.dropna(inplace=True)
```

```python
correl.head()
```

```python
correl.sort_values('correlation',ascending=False).head()
```

```python
#Which movies got number of ratings is more than 100 find the correlation with starwar movie.
correl[ratings['num_of_ratings']>100].sort_values('correlation',ascending=False).head()
```
